```python
import numpy as np
from sklearn.linear_model import LinearRegression

#Database
# x = Data, y = Target
x = [[1],[4],[7],[10],[13],[16],[19],[22],[25],[28]]
y = [2, 8, 14, 20, 26, 32, 38, 44, 50, 56]

regr = LinearRegression().fit(x,y)
regr.score(x,y)



#Data uji
predict = np.array([[31]])

#Menampilkan data prediksi
print ("Prediksi")
print ("Input = ", predict)
print ("Output = ", regr.predict(predict))
```
